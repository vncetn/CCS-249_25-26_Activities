### Updating ELIZA. Implementing RegEx on NLP. Perform the following preprocessing techniques using RegEx.

a. (10 points) Extract all of the words starting with an upper case letter
from the text:

```python
text = "Alice was beginning to get very tired of sitting by her sister on the bank, and of having nothing to do. Once or twice she had peeped into the book her sister was reading, but it had no pictures or conversations in it, \"and what is the use of a book,\" thought Alice, \"without pictures or conversations?"
```

```python
# PATTERN = [A-Z][a-z]*

uppercase_words = re.findall(pattern=r'\b[A-Z][a-z]*\b', string=text)
print(uppercase_words)
```

b. (10 points) Read the “melville-moby_dick.txt” into a Python program
and extract all of the instances of the word whale in said source.

```python
# PATTERN = (whales?)

moby_dick_path = r'C:\Users\Acer\Documents\Programs\CCS-249_25-26_Activities\BSCS 3B\DIANA\melville-moby_dick.txt'

with open(moby_dick_path, 'r', encoding='utf-8') as file:
    moby_dick_text = file.read()

whale_pattern = re.compile(r'\b(whales?)\b', re.IGNORECASE)

# whale_matches = re.findall(pattern=r'\b(whales?)\b'm string=moby_dick_text
whale_matches = whale_pattern.findall(moby_dick_text)

print(f"Number of whale instances: {len(whale_matches)}")
print(f"Whale instances {whale_matches}")
```

c. (10 points) Download the NLTK package into your local environment
and import webtext (from nltk.corpus import webtext). Load the
pirates.txt and extract all of the lines spoken by the character Jack
Sparrow.

```python
# PATTERN = ^(JACK SPARROW):\s*(.*)'

# nltk.download('webtext')
pirates_text = webtext.raw('pirates.txt')

pirates_pattern = re.compile(r'(JACK SPARROW):\s*(.*)', re.MULTILINE | re.IGNORECASE)
jack_spoken_lines = pirates_pattern.findall(pirates_text)

print(f"Number of lines spoken by Jack Sparrow: {len(jack_spoken_lines)}")
print(f"Lines spoken by Jack Sparrow:\n{jack_spoken_lines}")
```
