```python
import re
import nltk
from nltk.corpus import webtext
```

```python
# NLP Preprocessing with RegEx

def extract_uppercase_words(text):
    pattern = r'\b[A-Z]\w*'
    return re.findall(pattern, text)

text = """Alice was beginning to get very tired of sitting by her sister on the bank,
and of having nothing to do. Once or twice she had peeped into the book her sister
was reading, but it had no pictures or conversations in it, "and what is the use
of a book," thought Alice, "without pictures or conversations?" """

uppercase_words = extract_uppercase_words(text)


print("Uppercase words:")
print(uppercase_words)

# Total count
print("\nTotal number of words starting with uppercase:", len(uppercase_words))
```

```python

from google.colab import files

uploaded = files.upload()

filename = list(uploaded.keys())[0]

with open(filename, "r", encoding="utf-8") as file:
    text = file.read()

pattern_whale = r"\bwhales?\b"

whale_matches = re.findall(pattern_whale, text, flags=re.IGNORECASE)

print("Total occurrences of 'whale' and 'whales':", len(whale_matches))
print("\nAll matches:\n")
print(whale_matches)
```

```python

# Download required NLTK data
nltk.download('webtext')

# Load pirates.txt
pirates_text = webtext.raw('pirates.txt')


lines = pirates_text.split("\n")

pattern_jack = r"^JACK SPARROW:.*"

jack_sparrow_lines = [line for line in lines if re.match(pattern_jack, line)]

print("Total lines spoken by Jack Sparrow:", len(jack_sparrow_lines))
print("\nSample lines:\n")
for line in jack_sparrow_lines[:10]:
    print(line)
```
