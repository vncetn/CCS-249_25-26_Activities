```python
import re
from collections import defaultdict

question_history = defaultdict(int)
last_responses = {}

pattern_a = r"i\s+want\s+to\s+know\s+the\s+reasons\s+why\s+i\s+am\s+feeling\s+depressed"
response_a = "Why do you think you are feeling depressed all the time?"

pattern_b = r"i\s+am\s+feeling\s+stressed"
response_b = "What is causing you to feel stressed?"

pattern_c = r"my\s+feelings\s+towards\s+my\s+crush\s+are\s+invalidated"
response_c = "It's important that your feelings are validated. Can you tell me more about your crush?"

pattern_d = r"you\s+do(?:n't|\s+not)\s+understand\s+me"
response_d = "I understand that you feel I don't understand you. Help me understand you better."

pattern_e = r"i\s+can(?:'t|\s+not)\s+focus\s+on\s+my\s+studies"
response_e = "What might be preventing you from focusing on your studies?"

print("EXERCISE 1: ELIZA RegEx Patterns")
print("=" * 70)
print("\nPattern A:")
print(f"RegEx: {pattern_a}")
print("\nPattern B:")
print(f"RegEx: {pattern_b}")
print("\nPattern C:")
print(f"RegEx: {pattern_c}")
print("\nPattern D:")
print(f"RegEx: {pattern_d}")
print("\nPattern E:")
print(f"RegEx: {pattern_e}")
```

```python
def eliza_respond(user_input):
    """
    ELIZA-style response function with RegEx patterns
    Includes BONUS: Sarcastic response for repeated questions
    """
    user_lower = user_input.lower()
    
    if user_input in last_responses:
        question_history[user_input] += 1
        if question_history[user_input] > 1:
            return f"Oh wow, you're asking me the same thing again? I'm *so* excited to answer that once more!"
    else:
        question_history[user_input] = 1
    if re.search(pattern_a, user_lower, re.IGNORECASE):
        last_responses[user_input] = response_a
        return response_a
    elif re.search(pattern_b, user_lower, re.IGNORECASE):
        last_responses[user_input] = response_b
        return response_b
    elif re.search(pattern_c, user_lower, re.IGNORECASE):
        last_responses[user_input] = response_c
        return response_c
    elif re.search(pattern_d, user_lower, re.IGNORECASE):
        last_responses[user_input] = response_d
        return response_d
    elif re.search(pattern_e, user_lower, re.IGNORECASE):
        last_responses[user_input] = response_e
        return response_e
    else:
        return "Tell me more about that."

print("ELIZA Respond Function Defined!")
```

ELIZA Patterns (a-e)

```python
print("\n" + "="*70)
print("Pattern A Test:")
test_a = "I want to know the reasons why I am feeling depressed all the time."
print(f"Input: {test_a}")
print(f"Response: {eliza_respond(test_a)}")
```

```python
print("\n" + "="*70)
print("Pattern B Test:")
test_b = "I am feeling stressed."
print(f"Input: {test_b}")
print(f"Response: {eliza_respond(test_b)}")
```

### 2a

```python
print("="*60)
print("EXERCISE 2a: Words Starting with Uppercase")
print("="*60)

text_2a = """Alice was beginning to get very tired of sitting by her sister on the bank, and of having nothing to do. Once or twice she had peeped into the book her sister was reading, but it had no pictures or conversations in it, "and what is the use of a book," thought Alice, "without pictures or conversations?"""

pattern_2a = r'\b[A-Z][a-z]*\b'

uppercase_words = re.findall(pattern_2a, text_2a)

print(f"\nRegEx Pattern: {pattern_2a}")
print(f"\nText: {text_2a}")
print(f"\nWords found: {uppercase_words}")
print(f"Unique words: {list(set(uppercase_words))}")
```

### 2b

```python
print("\n" + "="*60)
print("EXERCISE 2b: Extract and Replace Whale Instances")
print("="*60)

file_path = r"C:\Users\ASUS\CCS-249_25-26_Activities\BSCS 3A\PAGAYON_ARTHUR_JOHN\melville-moby_dick.txt"

with open(file_path, 'r', encoding='utf-8') as file:
    moby_dick_text = file.read()

pattern_2b = r'\b[Ww]hale[s]?\b'

whale_instances = re.findall(pattern_2b, moby_dick_text)

print(f"\nRegEx Pattern: {pattern_2b}")
print(f"Total instances found: {len(whale_instances)}")
print(f"First 15 instances: {whale_instances[:15]}")
```

```python
modified_text = moby_dick_text
count = 0

def replace_first_n(match):
    global count
    count += 1
    if count <= 10:
        return "leviathan"
    else:
        return match.group(0)

modified_text = re.sub(pattern_2b, replace_first_n, moby_dick_text)

print(f"\nReplaced first 10 instances with 'leviathan'")

output_path = r"C:\Users\ASUS\CCS-249_25-26_Activities\BSCS 3A\PAGAYON_ARTHUR_JOHN\melville-moby_dick_modified.txt"
with open(output_path, 'w', encoding='utf-8') as file:
    file.write(modified_text)

print(f"Modified text saved to: {output_path}")
print(f"\nVerification - First 20 lines of modified text:")
print("\n".join(modified_text.split('\n')[:20]))
```
