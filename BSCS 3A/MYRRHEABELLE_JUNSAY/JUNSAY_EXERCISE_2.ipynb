## 1. Updating ELIZA
Modify the ELIZA python implementation shared to the class by performing the following:

### Add 5 RegEx patterns to capture:
1. `.*reasons why I am feeling depressed.*`
2. `.*feeling stressed.*`
3. `.*feelings towards my crush are invalidated.*`
4. `You (don\'t|do not) understand me`
5. `I (can\'t|cannot) focus on my studies`

### BONUS! Modify the code so that if the user tries to repeat the same question, ELIZA will answer sarcastically on repeating the question.

```python
previous_input = ""

def reflect(fragment):
    reflections = {
        "am": "are", "was": "were", "i": "you", "i'd": "you would",
        "i've": "you have", "i'll": "you will", "my": "your",
        "are": "am", "you've": "I have", "you'll": "I will",
        "your": "my", "yours": "mine", "you": "me", "me": "you"
    }
    words = fragment.lower().split()
    return ' '.join([reflections.get(word, word) for word in words])

def eliza_response(user_input):
    global previous_input

    # Sarcizm
    if user_input.lower() == previous_input.lower():
        return "Are you a broken record? I literally just answered that."

    previous_input = user_input

    patterns = [
        # Original
        (r"I need (.*)", "Why do you need {0}?"),
        (r"Why don’t you (.*)", "Do you really think I don't {0}?"),
        (r"I feel (.*)", "Tell me more about feeling {0}."),

        # New
        (r"I want to know the reasons why I am feeling depressed (.*)", "Depression is complex. Why do you feel {0} is a factor?"),
        (r"I am feeling stressed(.*)", "What do you think is causing this stress?"),
        (r"My feelings towards my crush are invalidated(.*)", "It is painful when your feelings are not recognized. Why do you feel they are invalidated?"),
        (r"You (don't|do not) understand me", "In what way do you feel I am failing to understand you?"),
        (r"I (can't|cannot) focus on my studies(.*)", "Studying is hard when your mind is elsewhere. What is distracting you {1}?")
    ]

    for pattern, response in patterns:
        match = re.match(pattern, user_input, re.IGNORECASE)
        if match:
            captured = match.group(match.lastindex) if match.lastindex else ""
            return response.format(reflect(captured), reflect(captured))

    return "Can you tell me more?"

test_queries = [
    "I am feeling stressed.",
    "You do not understand me.",
    "I cannot focus on my studies lately.",
    "I cannot focus on my studies lately."
]

print("ELIZA: Hello! How can I help you today?")
for q in test_queries:
    print(f"You: {q}")
    print(f"ELIZA: {eliza_response(q)}\n")
```

## 2. Implementing RegEx on NLP
Perform the following preprocessing techniques using RegEx.

### a. Extract all of the words starting with an upper case letter from the text.

**Pattern:** `\b[A-Z][a-z]*\b`


```python
text_alice = "Alice was beginning to get very tired of sitting by her sister on the bank, and of having nothing to do. Once or twice she had peeped into the book her sister was reading, but it had no pictures or conversations in it, \"and what is the use of a book,\" thought Alice, \"without pictures or conversations?"

pattern_a = r'\b[A-Z][a-z]*\b'
matches_a = re.findall(pattern_a, text_alice)

print(f"Words starting with uppercase: {matches_a}")
```

### b. Read the “melville-moby_dick.txt” into a Python program and extract all of the instances of the word Whale, Whales, whale and whales in said source. Then replace the first 10 instances with the word “leviathan”.

**Pattern:** `\b[Ww]hales?\b`

*Note:* *This strictly adheres to the instructions which specifically listed only four variants:* *Whale, Whales, whale, and whales*. *Fully capitalized instances like "WHALE" are excluded to ensure precision and strict compliance with the provided word list :D.*

```python
with open("melville-moby_dick.txt", "r") as file:
    moby_text = file.read()

whale_pattern = r'\b[Ww]hales?\b'

found_whales = re.findall(whale_pattern, moby_text)
print(f"Found {len(found_whales)} instances.")

new_text = re.sub(whale_pattern, "leviathan", moby_text, count=10)

print("\nModified Text:")
print(new_text[:900])
```

### c. Download the NLTK package into your local environment and import webtext (from nltk.corpus import webtext). Load the pirates.txt and extract all of the lines spoken by the character Jack Sparrow.

**Pattern:** `^JACK SPARROW:.*`

```python
pirates_raw = webtext.raw('pirates.txt')

jack_pattern = r'^JACK SPARROW:.*'
jack_lines = re.findall(jack_pattern, pirates_raw, re.MULTILINE)

print(f"Extracted {len(jack_lines)} lines for Jack Sparrow:\n")
for line in jack_lines:
    print(line)
```
