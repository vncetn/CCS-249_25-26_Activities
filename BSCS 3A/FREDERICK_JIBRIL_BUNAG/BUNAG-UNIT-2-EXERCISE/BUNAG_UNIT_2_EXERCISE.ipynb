## Exercise 2a: Extract words starting with uppercase letter

```python
import re

# Alice in Wonderland text
text = """Alice was beginning to get very tired of sitting by her sister on the bank, and of having nothing to do.  Once or twice she had peeped into the book her sister was reading, but it had no pictures or conversations in it, "and what is the use of a book," thought Alice, "without pictures or conversations?"""

# RegEx pattern to extract words starting with uppercase letter
pattern = r'\b[A-Z][a-z]*\b'

# Extract all words starting with uppercase
uppercase_words = re.findall(pattern, text)

print("Words starting with uppercase letter:")
print(uppercase_words)
print(f"\nTotal count: {len(uppercase_words)}")
```

## Exercise 2b: Extract and replace whale/Whale instances in Moby Dick

```python
import re

# Read the Moby Dick text file
with open('melville-moby_dick.txt', 'r', encoding='utf-8') as file:
    moby_dick_text = file.read()

# RegEx pattern to match Whale, Whales, whale, whales
pattern = r'\b[Ww]hales?\b'

# Extract all instances of whale/Whale
whale_instances = re.findall(pattern, moby_dick_text)

print(f"Total instances of whale/Whale found: {len(whale_instances)}")
print(f"First 20 instances: {whale_instances[:20]}")

# Replace the first 10 instances with "leviathan"
count = 0
def replace_first_ten(match):
    global count
    count += 1
    if count <= 10:
        return "leviathan"
    return match.group(0)

# Reset counter
count = 0
modified_text = re.sub(pattern, replace_first_ten, moby_dick_text)

print(f"\nFirst 10 instances replaced with 'leviathan'")

# Extract whale instances from modified text to verify
whale_instances_after = re.findall(pattern, modified_text)
print(f"Remaining whale/Whale instances after replacement: {len(whale_instances_after)}")

# Show a sample of the beginning to see the replacement
print("\nSample of modified text (first 1000 characters):")
print(modified_text[:1000])
```

## Exercise 2c: Extract Jack Sparrow's lines from pirates.txt

```python
import re
from nltk.corpus import webtext

# Load the pirates.txt file from webtext corpus
pirates_text = webtext.raw('pirates.txt')

# RegEx pattern to extract Jack Sparrow's lines
# The pattern looks for "JACK SPARROW:" or "JACK:" followed by the dialogue
pattern = r'(?:JACK SPARROW|JACK):\s*(.+?)(?=\n[A-Z]+(?:\s+[A-Z]+)*:|$)'

# Extract all lines spoken by Jack Sparrow
jack_sparrow_lines = re.findall(pattern, pirates_text, re.DOTALL)

print(f"Total lines spoken by Jack Sparrow: {len(jack_sparrow_lines)}")
print("\n" + "="*80)
print("First 10 lines spoken by Jack Sparrow:")
print("="*80)

for i, line in enumerate(jack_sparrow_lines[:10], 1):
    # Clean up the line by removing extra whitespace
    cleaned_line = ' '.join(line.split())
    print(f"\n{i}. {cleaned_line}")
```
