## 1. Updating ELIZA

```python
import re
import random

def reflect(fragment):
    """Reflects user input to make responses more natural."""
    reflections = {
        "am": "are",
        "was": "were",
        "i": "you",
        "i'd": "you would",
        "i've": "you have",
        "i'll": "you will",
        "my": "your",
        "are": "am",
        "you've": "I have",
        "you'll": "I will",
        "your": "my",
        "yours": "mine",
        "you": "me",
        "me": "you"
    }
    words = fragment.lower().split()
    return ' '.join([reflections.get(word, word) for word in words])

def eliza_response(user_input):
    """Generates ELIZA-style responses based on input."""
    patterns = [
        # Added Patterns
        # .a
        (r"I want to know the reasons why I am feeling depressed all the time", 
         "I am sorry to hear you feel that way. Have you spoken to anyone about this depression?"),
        # b.
        (r"I am feeling stressed", 
         "Stress can be difficult to handle. What do you think is the main cause?"),
        # c.
        (r"My feelings towards my crush are invalidated", 
         "It is painful when feelings are not acknowledged. Why do you feel they are invalidated?"),
        # d.
        (r"You (don't|do not) understand me", 
         "I am just a computer program, but I am trying. What do you wish I understood?"),
        # e.
        (r"I (can't|cannot) focus on my studies", 
         "Studying requires a clear mind. What is distracting you right now?"),

        # Original Patterns
        (r"I need (.*)", "Why do you need {0}?"),
        (r"Why don’t you (.*)", "Do you really think I don't {0}?"),
        (r"I feel (.*)", "Tell me more about feeling {0}.")
    ]
    
    for pattern, response in patterns:
        match = re.match(pattern, user_input, re.IGNORECASE)
        if match:
            # Check if the regex captured any groups (like {0} for reflection)
            if '{0}' in response:
                return response.format(reflect(match.group(1)))
            else:
                return response
    
    return "Can you tell me more?"

# Sarcastic responses
sarcastic_responses = [
    "Sulit sulit? Sulit sulit?",
    "Am I having a Deja Vu?",
    "Do you enjoy repeating yourself?",
    "Are you a chatbot built in the 60s too?",
    "*with unamused facial expression* I heard you the first time. Let's move on.",
    "Am I in a time loop?",
    "Am I talking to a machine?"
]

# Main Execution
print("ELIZA: Hello! How can I help you today?")

last_input = "" 

while True:
    user_input = input("You: ")
    
    if user_input.lower() in ["quit", "exit"]:
        print("ELIZA: Goodbye!")
        break
    
    # Check for repetition
    if user_input.strip().lower() == last_input.strip().lower() and last_input != "":
        print(f"ELIZA: {random.choice(sarcastic_responses)}")
        continue

    last_input = user_input
    
    print(f"ELIZA: {eliza_response(user_input)}")
```

## 2. Implementing RegEx on NLP

###  A. Extract all of the words starting with an upper case letter from the text.

```python
text = """Alice was beginning to get very tired of sitting by her sister on the bank,
and of having nothing to do. Once or twice she had peeped into the book
her sister was reading, but it had no pictures or conversations in it, "and
what is the use of a book," thought Alice, "without pictures or
conversations?"""

# Pattern to find words starting with an uppercase letter
pattern = r"\b[A-Z]\w*"

# Extract all matches
matches = re.findall(pattern, text)

print("Found words:", matches)
```

### B. Read the “melville-moby_dick.txt” into a Python program and extract all of the instances of the word Whale, Whales, whale and whales in said source. Then replace the first 10 instances with the word “leviathan”.

```python
def replace(match):
    """
    Determines the replacement word based on case and plurality.
    """
    word = match.group(0)
    
    # handle plural
    if word.lower().endswith('s'):
        replacement = "leviathans"
    else:
        replacement = "leviathan"    
    # Handle capitalization
    if word[0].isupper():
        replacement = replacement.capitalize()
    # Instances of all caps (WHALE) are not replaced as per instructions (extract only Whale, Whales, whale, whales)  
    return replacement

# Read the file
with open("melville-moby_dick.txt", "r") as file:
    content = file.read()

# Whale, Whales, whale, whales
pattern = r"\b[Ww]hales?\b"

# Replace the first 10 instances
new_content = re.sub(pattern, replace, content, count=10)

print(new_content)
```

### C. Download the NLTK package into your local environment and import webtext (from nltk.corpus import webtext). Load the pirates.txt and extract all of the lines spoken by the character Jack Sparrow.

```python
import nltk
from nltk.corpus import webtext

nltk.download('webtext')
pirates_content = webtext.raw('pirates.txt')

jack_lines = []

# Split the content into individual lines and check each one
for line in pirates_content.splitlines():
    # Check if the line starts with the Jack's tag
    if line.startswith('JACK SPARROW:'):
        jack_lines.append(line)

# --- Display Results ---
print(f"Found {len(jack_lines)} lines spoken by Jack Sparrow.")
print("-" * 50)
for line in jack_lines:
    print(line)
```
