```python
#Number 1 Updating the ELIZA
# ELIZA implementation in Python
# Example Generated via ChatGPT
import re

class Eliza:
    def __init__(self):
        self.last_input = None

        self.reflections = {
            "am": "are",
            "was": "were",
            "i": "you",
            "i'd": "you would",
            "i've": "you have",
            "i'll": "you will",
            "my": "your",
            "are": "am",
            "you've": "I have",
            "you'll": "I will",
            "your": "my",
            "yours": "mine",
            "you": "me",
            "me": "you",
            "can't": "can",
            "cannot": "can"
        }
        
        self.patterns = [
            (r"I need (.*)", "Why do you need {0}?"),
            (r"Why don’t you (.*)", "Do you really think I don't {0}?"),
            (r"I feel (.*)", "Tell me more about feeling {0}."),
            (r"I want to know the reasons why I am feeling depressed all the time", 
             "It sounds tough feeling depressed all the time. What do you think might be causing it?"),
            (r"I am feeling stressed", 
             "Stress can be overwhelming. What do you think is causing your stress?"),
            (r"My feelings towards my crush are invalidated", 
             "Why do you feel your feelings are being invalidated?"),
            (r"You don’t understand me|You do not understand me", 
             "It seems like you feel misunderstood. Can you explain more?"),
            (r"I can’t focus on my studies|I cannot focus on my studies", 
             "Focusing can be hard sometimes. What is distracting you?")
        ]

    def reflect(self, fragment):
        words = fragment.lower().split()
        return ' '.join([self.reflections.get(word, word)for word in words])

    def respond(self, user_input):
        if self.last_input and user_input.lower() == self.last_input.lower():
            return "Really? You're asking that again? Interesting."
        
        self.last_input = user_input

        for pattern, response in self.patterns:
            match = re.match(pattern, user_input, re.IGNORECASE)
            if match:
                if match.groups():
                    return response.format(self.reflect(match.group(1)))
                else:
                    return response
            return "Can you tell me more?"
    
    def chat(self):
        print("ELIZA: Hello! How can I help you today?")
        while True:
            user_input = input("You: ")
            if user_input.lower() in ["quit", "exit"]:
                print("ELIZA: Goodbye!")
                break
            print(f"ELIZA: {self.respond(user_input)}")

if __name__ == "__main__":
    eliza_bot = Eliza()
    eliza_bot.chat()

# def reflect(fragment):
#     """Reflects user input to make responses more natural."""
#     reflections = {
#         "am": "are",
#         "was": "were",
#         "i": "you",
#         "i'd": "you would",
#         "i've": "you have",
#         "i'll": "you will",
#         "my": "your",
#         "are": "am",
#         "you've": "I have",
#         "you'll": "I will",
#         "your": "my",
#         "yours": "mine",
#         "you": "me",
#         "me": "you"
#     }
#     words = fragment.lower().split()
#     return ' '.join([reflections.get(word, word) for word in words])

# def eliza_response(user_input):
#     """Generates ELIZA-style responses based on input."""
#     patterns = [
#         (r"I need (.*)", "Why do you need {0}?"),
#         (r"Why don’t you (.*)", "Do you really think I don't {0}?"),
#         (r"I feel (.*)", "Tell me more about feeling {0}.")
#     ]
    
#     for pattern, response in patterns:
#         match = re.match(pattern, user_input, re.IGNORECASE)
#         # print(match)
#         if match:
#             print(match.group(1)) # captures the substring after the pattern
#             return response.format(reflect(match.group(1)))
    
#     return "Can you tell me more?"

# print("ELIZA: Hello! How can I help you today?")
# while True:
#     user_input = input("You: ")
#     if user_input.lower() in ["quit", "exit"]:
#         print("ELIZA: Goodbye!")
#         break
#     print(f"ELIZA: {eliza_response(user_input)}")
```

```python
#2.b
import re

with open("datasets/melville-moby_dick.txt", "r", encoding="utf-8") as file:
    text = file.read()

pattern = r'\b[Ww]hale[s]?\b' #Whale, Whales, whale, whales

matches = re.findall(pattern, text)
print(f"Total matches found: {len(matches)}")
print(f"First 20 matches: {matches[:20]}")

def replace_first_n(pattern, repl, string, n):
    count = 0
    def _repl(match):
        nonlocal count
        if count < n:
            count += 1
            return repl
        return match.group(0)
    return re.sub(pattern, _repl, string)

updated_text = replace_first_n(pattern, "leviathan", text, 10)

print("\nFirst 500 characters of updated text:\n")
print(updated_text[:2000])

```
